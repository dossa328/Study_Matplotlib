# 3.1. 여러 개의 그래프 그리기

### 3.1.1 곡선 그리기

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.01)
plt.plot(t1, f(t1))
plt.show()
```

* 위 그래프는 $ y = exp(-x) * cos(2\pi x) $ 를 x 가 0 ~ 5 까지 범위일 때를 그린것이다
* 만약 `t1 = np.arange(0.0, 5.0, 0.5)` 와 같이 간격을 0.5로 두면 다음과 같은 그래프를 구할 수 있다

```python
t1 = np.arange(0.0, 5.0, 0.5)
plt.plot(t1, f(t1))
plt.show()
```

* 간격을 0.01과 같이 하면 출력되는 간격이 좁아져서 부드러운 곡선을 구할 수 있다.

### 3.1.2 곡선 그래프 2개를 하나의 셀에 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

def g(t):
    return np.sin(np.pi*t)

t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.01)

plt.plot(t1, f(t1), t2, g(t2), 'r-')

plt.show()
```

### 3.1.3. 분할해서 그리기
* subplot을 사용하면 된다

#### 상하 2분할 그래프

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

def g(t):
    return np.sin(np.pi*t)

t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.01)

plt.subplot(211)
plt.plot(t1, f(t1), 'r-')

plt.subplot(212)
plt.plot(t2, g(t2), 'b-')

plt.show()
```

#### 4분할 그래프

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

def g(t):
    return np.sin(np.pi*t)

t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.01)

plt.subplot(221)
plt.plot(t1,f(t1))

plt.subplot(222)
plt.plot(t2, g(t2))

plt.subplot(223)
plt.plot(t1, f(t1), 'r-')

plt.subplot(224)
plt.plot(t2, g(t2), 'r-')

plt.show()
```

#### 여러개의 창(figure)에 그래프 그리기

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

def g(t):
    return np.sin(np.pi*t)

t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.01)

plt.figure(1)
plt.plot(t1, f(t1))

plt.figure(2)
plt.plot(t2, g(t2))

plt.show()
```
